# Intronduction
We recognize the works of artists through their unique style, such as color choices or brush strokes. The “je ne sais quoi” of artists like Claude Monet can now be imitated with algorithms thanks to generative adversarial networks (GANs). In this getting started competition, you will bring that style to your photos or recreate the style from scratch!

Computer vision has advanced tremendously in recent years and GANs are now capable of mimicking objects in a very convincing way. But creating museum-worthy masterpieces is thought of to be, well, more art than science. So can (data) science, in the form of GANs, trick classifiers into believing you’ve created a true Monet? That’s the challenge you’ll take on!

### The Challenge:
A GAN consists of at least two neural networks: a generator model and a discriminator model. The generator is a neural network that creates the images. For our competition, you should generate images in the style of Monet. This generator is trained using a discriminator.

The two models will work against each other, with the generator trying to trick the discriminator, and the discriminator trying to accurately classify the real vs. generated images.

Your task is to build a GAN that generates 7,000 to 10,000 Monet-style images.

# Dataset
The dataset contains four directories: __monet_tfrec__, __photo_tfrec__, __monet_jpg__, and __photo_jpg__. The __monet_tfrec__ and __monet_jpg__ directories contain the same painting images, and the __photo_tfrec__ and __photo_jpg__ directories contain the same photos.

We recommend using TFRecords as a Getting Started competition is a great way to become more familiar with a new data format, but JPEG images have also been provided.

The monet directories contain Monet paintings. Use these images to train your model.

The photo directories contain photos. Add Monet-style to these images and submit your generated jpeg images as a zip file. Other photos outside of this dataset can be transformed but keep your submission file limited to 10,000 images.

Note: Monet-style art can be created from scratch using other GAN architectures like DCGAN. The submitted image files do not necessarily have to be transformed photos.

```python
### Prepare Pipeline
AUTOTUNE = tf.data.AUTOTUNE
```

```python
# Read Dataset
def read_tfrecord(example):
    feature_description = {
        'image_name': tf.io.FixedLenFeature([], tf.string),
        'image' : tf.io.FixedLenFeature([], tf.string),
        'target': tf.io.FixedLenFeature([],tf.string)
    }
    parsed_record = tf.io.parse_single_example(example, feature_description)
    image = tf.image.decode_jpeg(parsed_record['image'], channels=3)
    return image

def read_file(DATA_PATH):
    FILENAMES = tf.io.gfile.glob(DATA_PATH)  
    raw_dataset = tf.data.TFRecordDataset(FILENAMES)
    dataSize = sum(1 for _ in tf.data.TFRecordDataset(FILENAMES))
    dataset = raw_dataset.map(read_tfrecord,num_parallel_calls = AUTOTUNE)
    return dataset,dataSize

def read_imgSize(dataset):
    height = set()
    width = set()
    channel = set()
    for i, img in enumerate(dataset):
        img_size = tf.shape(img).numpy()
        height.add(img_size[0])
        width.add(img_size[1])
        channel.add(img_size[2])
    return height, width, channel
```

```python
DATA_PATH = "data/"
dataset_monet,dataSize_monet = read_file(DATA_PATH + 'monet_tfrec/*.tfrec')
dataset_real,dataSize_real = read_file(DATA_PATH + 'photo_tfrec/*.tfrec')
print("number of Monet paints:",dataSize_monet)
print("number of real img:", dataSize_real)

# All the image are of the same size
print("Monet img size:", read_imgSize(dataset_monet))
print("Real img size", read_imgSize(dataset_real))

IMG_WIDTH, IMG_HEIGHT, IMG_CHANNEL = 256, 256, 3
```

```python
# Prepare Train, Validation and Test dataset
def split_data(dataset, train_ratio, valid_ratio, test_ratio):
    trainSize = int(train_ratio * dataSize_monet)
    validSize = int(valid_ratio * dataSize_monet)
    testSize = int(test_ratio * dataSize_monet)
    
    train_data = dataset.take(trainSize)
    
    valid_data = dataset.skip(trainSize)
    valid_data = dataset.take(validSize)
    
    test_data = dataset.skip(trainSize + validSize)
    
    return train_data, valid_data, test_data

trainData_monet, validData_monet, testData_monet = split_data(dataset_monet, 0.8, 0.8, 0.1)
trainData_real, validData_real, testData_real = split_data(dataset_real, 0.8, 0.8, 0.1)
```

# Data Preprocessing

Image Normalization + Image Augmentation

```python
def data_normalize(image):
    # image normalization
    image = tf.cast(image, tf.float32)
    image = tf.image.per_image_standardization(image)
    return image
    
def data_augment(image):
    # Rezie the image and randomly chop
    image = tf.image.resize(image, size = [270,  270], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    image = tf.image.random_crop(image, size=[IMG_HEIGHT, IMG_WIDTH, IMG_CHANNEL])
    #left-right flip
    image = tf.image.random_flip_left_right(image)
    return image

def data_process(dataset, flag_aug = False):
    if flag_aug:
        dataset = dataset.map(data_augment, num_parallel_calls = AUTOTUNE)
    dataset = dataset.map(data_normalize, num_parallel_calls = AUTOTUNE)
    return dataset
```

```python
BATCH_SIZE = 64
```

```python
#Process Monet train, valid, test data
trainData_monet = data_process(trainData_monet, flag_aug = True).cache()
validData_monet = data_process(validData_monet, flag_aug = False).cache()
testData_monet = data_process(testData_monet, flag_aug = False).cache()


#Process Real train, valid, test data
trainData_real = data_process(trainData_real, flag_aug = True).cache()
validData_real = data_process(validData_real, flag_aug = False).cache()
testData_real = data_process(testData_real, flag_aug = False).cache()
```
